```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
import time
```

```python
trainingData = pd.read_csv('flatPrices/train.csv')
testingData = pd.read_csv('flatPrices/test.csv')


X_train = trainingData.drop(columns='resale_price')
X_train = X_train.iloc[:-1]
y_train = trainingData['resale_price']
y_train = y_train.iloc[:-1]

X_test = testingData.drop(columns='resale_price')
X_test = X_test.iloc[:-1]
y_test = testingData['resale_price']
y_test = y_test.iloc[:-1]
```

```python
from sklearn.linear_model import LinearRegression 
LR = LinearRegression()
```

```python
start_time = time.time()
LR.fit(X_train, y_train)
elapsed_time=time.time()-start_time
```

```python
#print('Time taken to train model classically :', round(elapsed time))
```

```python
y_pred = LR.predict(X_test)
```

```python
plt.scatter(y_test,y_pred)
```

```python
sns.distplot((y_test-y_pred),bins=50);
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
