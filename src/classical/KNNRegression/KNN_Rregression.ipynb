```python
feature_number = 20
test_size = 0.3
y_feature = 'Class'
```

```python
import pandas as pd
import numpy as np
import time
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(f'../../data/credit/{feature_number}F/data.csv')
X = data.drop(columns=y_feature,axis=1)
y = data[y_feature]
data.head()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
scaled_features = scaler.transform(X)
```

```python
df_feat = pd.DataFrame(scaled_features,columns=X.columns)
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,y,
                                                    test_size=test_size)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
start_time = time.time()
knn.fit(X_train,y_train)
elapsed_time = time.time() - start_time
print(f'Time spent to fit: {elapsed_time}' )
```

```python
start_time = time.time()
y_pred = knn.predict(X_test)
elapsed_time = time.time() - start_time
elapsed_time
print(f'Time spent to predict: {elapsed_time}' )
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
start_time = time.time()
score = knn.score(X_test, y_test)
elapsed_time = time.time() - start_time
print(f'score : {score} \ntime to score: {elapsed_time}' )
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
