```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MaxAbsScaler
%matplotlib inline
```

```python
data = pd.read_csv('./sam.csv')
data.sort_values(by=['feature'])
```

```python
sns.set_palette("GnBu_d")
sns.set_style('whitegrid')
data.columns
```

```python
cpu16 = data[data['cpu'] == 1.6]
cpu25 = data[data['cpu'] == 2.5]
cpu22 = data[data['cpu'] == 2.2]
cpu3 = data[data['cpu'] == 3]
qc = data[data['cpu'] == 7]

plt.plot(cpu16['feature'],cpu16['score'],  label='1.6GHz')
plt.plot(cpu22['feature'],cpu22['score'],  label='2.2GHz')
plt.plot(cpu25['feature'],cpu25['score'],  label='2.5GHz')
plt.plot(cpu3['feature'],cpu3['score'],  label='3GHz')
plt.plot(qc['feature'],qc['score'], '-.', label='Quantum')

plt.legend()
plt.ylabel('Error')
plt.xlabel('number of features/ columns')


```

```python
plt.plot(cpu16['feature'],cpu16['MAE'],  label='1.6GHz')
plt.plot(cpu22['feature'],cpu22['MAE'],  label='2.2GHz')
plt.plot(cpu25['feature'],cpu25['MAE'],  label='2.5GHz')
plt.plot(cpu3['feature'],cpu3['MAE'],  label='3GHz')
plt.plot(qc['feature'],qc['MAE'], '-.',color='red', label='Quantum')

plt.legend()
plt.ylabel('MAE')
plt.xlabel('number of features/ columns')
```
