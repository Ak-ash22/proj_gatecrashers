## hdb flat prices 1990 2021 march

__ALL Prices 1990-2021 Mar.csv__

The largest file by far, ALL Prices 1990-2021 Mar.csv, contains over 800k rows of transactions of Singapore Housing Development Board (HDB) resale flats. As the name implies, BTO, SERS, HUDC and private housing are not included, though resale DBSS flat transactions are treated as ordinary HDB flat transactions. Many of the columns in the file are calculated columns or mapped columns (based on supplementary information) like the CPI index and lease percentage columns. For the full metadata/glossary of how I derived these terms, see the bottom of this description and/or the file and column descriptions.ons.

```python
import pandas as pd
```

```python
data = pd.read_csv('../data/row_data/ALL_ Prices1990-2021mar.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
data = data.drop(columns=['month',
                          'town_dummy',
                         'street_name',
                         'address',
                         'storey_range',
                         'storey',
                         'lease_start',
                         'lease_rem',
                         'price_psm',
                         'price_psm_yearly',
                         'Core CPI',
                         'latitude',
                         'longitude',
                         'price cpi_adj',
                         'price_psm cpi_adj',
                         'bala lease pct',
                         'price lease_adj implied',
                         'price_psm lease_adj implied',
                         'price_psm cpi_lease_adj implied',
                         'year_gni',
                         'price cpi_lease_adj implied',
                         'flat_model'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
X = data.drop(columns='resale_price')
y = data['resale_price']
```

```python
LE = LabelEncoder()
```

```python
data['flat_type']
```

```python
X['town'] = LE.fit_transform(X['town'])
X['flat_type'] = LE.fit_transform(X['flat_type'])
X['block'] = LE.fit_transform(X['block'])
```

```python
X['flat_type']
```

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
X[['area_sqm']] = MaxAbsScaler().fit_transform(X[['area_sqm']])
X[['year']] = MaxAbsScaler().fit_transform(X[['year']])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=1111)
```

```python
X_train
```

```python
#I USED THE FOLLOWING TO CREATE THE DATA
X_train.to_csv('../data/flat_data/5F/X_train_5F.csv', encoding='utf-8', index=False)
y_train.to_csv('../data/flat_data/5F/y_train_5F.csv', encoding='utf-8', index=False)
X_test.to_csv('../data/flat_data/5F/X_test_5F.csv', encoding='utf-8', index=False)
y_test.to_csv('../data/flat_data/5F/y_test_5F.csv', encoding='utf-8', index=False)
```
