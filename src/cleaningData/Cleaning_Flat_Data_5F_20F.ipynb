## hdb flat prices 1990 2021 march

__ALL Prices 1990-2021 Mar.csv__

The largest file by far, ALL Prices 1990-2021 Mar.csv, contains over 800k rows of transactions of Singapore Housing Development Board (HDB) resale flats. As the name implies, BTO, SERS, HUDC and private housing are not included, though resale DBSS flat transactions are treated as ordinary HDB flat transactions. Many of the columns in the file are calculated columns or mapped columns (based on supplementary information) like the CPI index and lease percentage columns. For the full metadata/glossary of how I derived these terms, see the bottom of this description and/or the file and column descriptions.ons.

```python
import pandas as pd
```

```python
data = pd.read_csv('../data/row_data/ALL_ Prices1990-2021mar.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data20 = data.drop(columns=['month',
                          'town_dummy',
                         'street_name',
                         'address',
                         'Core CPI',
                         'year_gni',
                         'price cpi_lease_adj implied'
                         ])

data5 = data.drop(columns=['month',
                          'town_dummy',
                         'street_name',
                         'address',
                         'storey_range',
                         'storey',
                         'lease_start',
                         'lease_rem',
                         'price_psm',
                         'price_psm_yearly',
                         'Core CPI',
                         'latitude',
                         'longitude',
                         'price cpi_adj',
                         'price_psm cpi_adj',
                         'bala lease pct',
                         'price lease_adj implied',
                         'price_psm lease_adj implied',
                         'price_psm cpi_lease_adj implied',
                         'year_gni',
                         'price cpi_lease_adj implied',
                         'flat_model'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
data20['town'] = LE.fit_transform(data20['town'])
data20['flat_type'] = LE.fit_transform(data20['flat_type'])
data20['block'] = LE.fit_transform(data20['block'])
data20['storey_range'] = LE.fit_transform(data20['storey_range'])
data20['flat_model'] = LE.fit_transform(data20['flat_model'])


data5['town'] = LE.fit_transform(data5['town'])
data5['flat_type'] = LE.fit_transform(data5['flat_type'])
data5['block'] = LE.fit_transform(data5['block'])
```

```python
data.info()
```

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
data20[['longitude']] = MaxAbsScaler().fit_transform(data20[['longitude']])
data20[['area_sqm']] = MaxAbsScaler().fit_transform(data20[['area_sqm']])
data20[['year']] = MaxAbsScaler().fit_transform(data20[['year']])
data20[['price lease_adj implied']] = MaxAbsScaler().fit_transform(data20[['price lease_adj implied']])
data20[['price_psm lease_adj implied']] = MaxAbsScaler().fit_transform(data20[['price_psm lease_adj implied']])
data20[['price_psm']] = MaxAbsScaler().fit_transform(data20[['price_psm']])
data20[['lease_start']] = MaxAbsScaler().fit_transform(data20[['lease_start']])
data20[['price_psm_yearly']] = MaxAbsScaler().fit_transform(data20[['price_psm_yearly']])
data20[['price cpi_adj']] = MaxAbsScaler().fit_transform(data20[['price cpi_adj']])
data20[['bala lease pct']] = MaxAbsScaler().fit_transform(data20[['bala lease pct']])
data20[['price_psm cpi_adj']] = MaxAbsScaler().fit_transform(data20[['price_psm cpi_adj']])
data20[['price_psm cpi_lease_adj implied']] = MaxAbsScaler().fit_transform(data20[['price_psm cpi_lease_adj implied']])



data5[['area_sqm']] = MaxAbsScaler().fit_transform(data5[['area_sqm']])
data5[['year']] = MaxAbsScaler().fit_transform(data5[['year']])
data5[['block']] = MaxAbsScaler().fit_transform(data5[['block']])
data5[['flat_type']] = MaxAbsScaler().fit_transform(data5[['flat_type']])
```
